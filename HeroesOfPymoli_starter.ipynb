## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,10000]
group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

gender_demographics_df["Age Ranges"] = pd.cut(gender_demographics_df["Age"], bins, labels=group)

age_demographics_totals = gender_demographics_df["Age Ranges"].value_counts()
age_demographics_percents = age_demographics_totals / total_players * 100

age_demographics = pd.DataFrame({"Total Count": age_demographics_totals, "Percent of Players": age_demographics_percents})
age_demographics = age_demographics.sort_index()

age_demographics.round(2)
```
